```python
import seaborn as sns
import matplotlib.pyplot as plt
import random

for i in range(0,10):
    df_melt = df.melt(id_vars=['benchmark', 'Task group', 'agent'], 
                    value_vars=['max DOM k-tokens', 'max Pruned DOM k-tokens', 'max AXTree k-tokens', 'episode DOM k-tokens', 'episode Pruned DOM k-tokens', 'episode AXTree k-tokens'], 
                    var_name='experience', value_name='value')
    #take all value and multiply by 1000
    df_melt['value'] = df_melt['value'] * 1000

    df_melt[['type', 'obs']] = df_melt['experience'].str.split(' ', n=1, expand=True)
    #drop columns
    df_melt = df_melt.drop(columns=['experience'])
    #rearraange columns
    df_melt = df_melt[['benchmark', 'Task group', 'agent', 'type', 'obs', 'value']]
    sorted_obs = df_melt.groupby('obs')['value'].mean().sort_values().index

    print(df_melt.head())

    #drop lines where obs contains 'pruned'
    df_melt = df_melt[~df_melt['obs'].str.contains('Pruned')]

    #keeping only max 
    df_melt = df_melt.loc[df_melt['type'] == 'max']
    #df_melt = df_melt.loc[df_melt['agent'] == 'base_agent']
    df_melt['Task group'] = df_melt['benchmark'] + ' ' + df_melt['Task group']
    #remove lines where task group = "workarena [ALL TASKS]"
    df_melt = df_melt[df_melt['Task group'] != "workarena [ALL TASKS]"]
    # Create the bar plot
    # Create a sorted list of 'obs' values based on the mean 'value'

    # Change the order of 'obs' in df_melt based on sorted_obs
    sns.set_style("darkgrid")
    np.random.seed(random.randint(0, 100))

    # Create a color palette with a random seed
    palette = sns.color_palette('colorblind', n_colors=len(df_melt['Task group'].unique()))
    random.shuffle(palette)
    palette_grey = sns.color_palette('grey', n_colors=1)
    print(df_melt.head())
    # Sort df_melt by 'obs'
    sorted_task_group = df_melt.groupby(['Task group'])['value'].mean().sort_values().index.get_level_values('Task group')
    figure = plt.figure(figsize=(10, 5))
    df_b = df_melt.loc[df_melt['agent'] == 'base_agent']
    sns.barplot(x='obs', y='value', hue='Task group', hue_order=sorted_task_group, data=df_b, order=None, palette=palette, ci=None)

    # Create the bar plot with sorted 'obs' categories
    df_f = df_melt.loc[df_melt['agent'] == 'full_agent']
    sns.barplot(x='obs', y='value', hue='Task group', hue_order=sorted_task_group, data=df_f, order=None, palette=[(0.75, 0.75, 0.7)], ci=None)

    df_b = df_melt.loc[df_melt['agent'] == 'base_agent']
    sns.barplot(x='obs', y='value', hue='Task group', hue_order=sorted_task_group, data=df_b, order=None, palette=palette, ci=None)

    plt.ylabel('Max number of token per step (log scale)', fontsize=16)
    plt.xlabel('')
    #custom x tick labels
    plt.xticks([0, 1], ['Document Object Model', 'Accessibility Tree'] , fontsize=16)
    plt.yticks(fontsize=16)

    #y axis log
    plt.yscale('log')

    #put legend on top two rows aligned legend word on the right

    legend1 = plt.legend(loc='upper center', fontsize=16, bbox_to_anchor=(0.5, 1.24), ncol=4, labels=["MiniWoB", "knowledge", "menu", "service catalog","list-sort" , "form", "list-filter"])

    plt.savefig(f'wa_complexity_{i}.pdf', dpi=300, bbox_inches='tight')
    plt.savefig(f'wa_complexity_{i}.png', dpi=300, bbox_inches='tight')



```
